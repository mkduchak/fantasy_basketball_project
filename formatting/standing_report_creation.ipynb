```python
import pandas as pd
import datetime
import os
import matplotlib as pyplot
%matplotlib inline
```

```python
yesterday = (datetime.datetime.today() -  datetime.timedelta(1)).strftime('%Y-%m-%d')
preyesterday = (datetime.datetime.today() -  datetime.timedelta(2)).strftime('%Y-%m-%d')
```

```python
standings_filepath = '../scraper/data/standings/'
standings_files = os.listdir(standings_filepath)[:-1]
standings_fileslist = []
for file in standings_files:
    standings_fileslist.append(standings_filepath + file)
```

```python
full_standing_df = pd.DataFrame()
for file in standings_fileslist:
    new_data = pd.read_csv(file, index_col=0)
    full_standing_df = full_standing_df.append(new_data)
```

```python
full_standing_df['rbs/g'] = full_standing_df['rebounds'] / full_standing_df['games_played']
full_standing_df['ast/g'] = full_standing_df['assists'] / full_standing_df['games_played']
full_standing_df['stl/g'] = full_standing_df['steals'] / full_standing_df['games_played']
full_standing_df['blk/g'] = full_standing_df['blocks'] / full_standing_df['games_played']
full_standing_df['pts/g'] = full_standing_df['points'] / full_standing_df['games_played']
```
