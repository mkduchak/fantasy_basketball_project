```python
import os
import pandas as pd
```

```python
detail_filepath = '../scraper/data/dailyteams/'
detail_files = os.listdir(detail_filepath)[:-1]
detail_fileslist = []
for file in detail_files:
    if file[-4:] == ".csv":
        detail_fileslist.append(detail_filepath + file)
```

```python
full_detail_df = pd.DataFrame()
for file in detail_fileslist:
    new_data = pd.read_csv(file)
    full_detail_df = full_detail_df.append(new_data)
```

```python
played_df = full_detail_df[(full_detail_df['MIN'] > 0) & (full_detail_df['pos']!='Bench')]
```
